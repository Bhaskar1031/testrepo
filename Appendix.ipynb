```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import wget
from folium.plugins import MarkerCluster
from folium.plugins import MousePosition
from folium.features import DivIcon
```

```python
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
```

```python
df_updated = df.groupby(['Orbit']).mean()
```

```python
sns.barplot(x=df_updated.index.values, y="Reused", data=df_updated)
plt.ylabel("Reused",fontsize=20)
plt.xlabel("Orbit",fontsize=20)
plt.show()
```

```python
df_updated_2 = df.groupby(['Orbit']).mean()
sns.barplot(x=df_updated_2.index.values, y="ReusedCount", data=df_updated_2)
plt.ylabel("Average Reused Count",fontsize=20)
plt.xlabel("Orbit",fontsize=20)
plt.show()
```

```python
df_graph=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_dash.csv")
```

```python
df_graph_updated = df_graph.groupby(['Booster Version Category']).mean()
```

```python
sns.barplot(x=df_graph_updated.index.values, y="class", data=df_graph_updated)
plt.ylabel("Class",fontsize=20)
plt.xlabel("Booster version Category",fontsize=20)
plt.show()
```

```python
sns.barplot(x=df_graph_updated.index.values, y="Payload Mass (kg)", data=df_graph_updated)
plt.ylabel("Payload Mass (kg)",fontsize=20)
plt.xlabel("Booster version Category",fontsize=20)
plt.show()
```
